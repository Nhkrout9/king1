```python
def knn(x,y,query_point,k=5):
  #define an array
  vals=[]
  #which will store distance value
  m=x.shape[0]
  #total number of points 
  for i in range(m):
    d=dist(query_point,x[i])
    vals.append((d,y[i]))
  vals=vals[:k]
  vals=np.array(vals)
  new_vals=np.unique(vals[:,1],return_counts=True)
  #print(new_vals)
  
  index=new_vals[1].argmax()
  #print(index)
  
  pred=new_vals[0][index]
  #print(pred)
  return pred
knn(x,y,query_point)  
#print(t)
```
