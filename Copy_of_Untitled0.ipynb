```python
## assignment 3 loop
input_size=int(input())
l=[]
for i in range(input_size):
  p=int(input())

  l.append(p)
  print("\n ")    

for i in range(input_size):
  p=l[i]
  for j in range(2,p):
    if(p%j==0):
      break
    else:
      if(j==(p-1)):
        print("\n ")    
        print(l[i],end="  ")
```
